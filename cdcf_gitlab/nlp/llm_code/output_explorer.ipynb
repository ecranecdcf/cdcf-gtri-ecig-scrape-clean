## Load up CSV

For now, we have two options, use comment/uncomment to look at Vape WH vs. CS Vape.

```python
csv_file = 'datasets/output/processed_output/10_29_2_vapewh_processed.csv' #VAPE WH
# csv_file = 'datasets/output/processed_output/10_29_2_csvape_processed.csv' #CS VAPE
df = pd.read_csv(csv_file)
```

```python
cand_cols = [ 'title', 'link',   'FINAL_Nic_Free','tfn_proc_llm', 'cbd_proc_llm', 'product_type_proc_llm',
              'FINAL_E-liquid contents', 'FINAL_Nicotine_Levels', 'FINAL_Nic_level_1',
       'FINAL_Nic_level_2', 'FINAL_Nic_level_3', 'FINAL_Nic_level_4',
       'FINAL_Nic_level_5', 'FINAL_Nic_level_6', 'FINAL_Nic_level_7',
    ]

final_cols = list()
for c in cand_cols:
    if c in df.columns:
        final_cols.append(c)
df = df[final_cols]
df
```
